```python
import sage.all as sage

# Define the prime number p
p = 23

# Find a primitive root modulo p
g = sage.primitive_root(p)

# Print the generator of the primitive root
print("The generator of a primitive root modulo", p, "is:", g)
```

```python
# Generate and print the sequence
sequence = [sage.power_mod(g, i, p) for i in range(1, p)]
print("The sequence generated by the primitive root is:", sequence)
```

```python
# Importing SageMath
import sage.all as sage

# Define the prime number p and a primitive root modulo p
p = 23
g = sage.primitive_root(p)

# Alice and Bob choose their private keys (random numbers smaller than p)
alice_private_key = 6
bob_private_key = 15

# Alice and Bob compute their public keys
alice_public_key = sage.power_mod(g, alice_private_key, p)
bob_public_key = sage.power_mod(g, bob_private_key, p)

# Print Alice's and Bob's public keys
print("Alice's public key:", alice_public_key)
print("Bob's public key:", bob_public_key)

# Alice and Bob compute the shared secret key using each other's public keys
alice_shared_key = sage.power_mod(bob_public_key, alice_private_key, p)
bob_shared_key = sage.power_mod(alice_public_key, bob_private_key, p)

# Print the shared secret key
print("Shared secret key (computed by Alice):", alice_shared_key)
print("Shared secret key (computed by Bob):", bob_shared_key)
```
