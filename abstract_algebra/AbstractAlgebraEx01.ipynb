```python
# Rings both addition and multiplication (no square brackets needed)
R=Integers(10)
R.list() # group under addition
```

```python
# 2 is not there because 2 and 10 do not have a GCD of 1, so no power of 2 will end up giving 1
R.list_of_elements_of_multiplicative_group() # Group under multiplication
```

```python
# Make is the order under addition of the member 2 (no square brackets needed)
c = R(2)
c.order()
```

```python
c=R(3)
print(f"Additive order of {c.order()}") #additive order
k = c.multiplicative_order()
print(f"Multiplicative order of {k}") #multiplicative order
# Print the multiplicative group
for n in range(k*2):
    print(c^n)
```

```python
# Print the additive group
k = c.order()
# Print the additive group
for n in range(k*2):
    print(n*c)
```
